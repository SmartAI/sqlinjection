```python
import pandas as pd
df1 = pd.read_csv('1.csv')
df2 = pd.read_csv('2.csv')
df3 = pd.read_csv('3.csv')
```

```python
total = pd.concat([df1, df2, df3])
total.info()
```

```python
# check labels
from matplotlib import pyplot as plt
%matplotlib inline
total.label.value_counts()
```

```python
pos = total[total['label'] == 1]
pos
```

```python
# 对数据做预处理

total = total.drop(['n', 'id'], 1)
total['query'] = total['raw_query'].apply(lambda x : re.sub(pattern, '', x))
total = total.drop('raw_query', 1)
total
```

```python
# 构造3-gram特征
matrix = total.as_matrix()
X = matrix[:, 1]
Y = matrix[:, 0]
Y = np.array(Y, dtype=np.int)

def get_gram(query):
    n = 3
    ngrams = []
    for i in range(0, len(query) - n):
        ngrams.append(query[i:i+n])
    return ngrams
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cross_validation import train_test_split
vectorizer = TfidfVectorizer(tokenizer=get_gram)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score
lgs = RandomForestClassifier()
X_train = vectorizer.fit_transform(X_train)
lgs.fit(X_train, y_train)
X_test = vectorizer.transform(X_test)
y_predict = lgs.predict(X_test)
from sklearn.metrics import f1_score
f1_score(y_test, y_predict)
```
